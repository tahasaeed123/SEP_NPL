```python
import pandas as pd
import nltk

df = pd.read_csv('/Users/tahasaeed/Documents/GreenQuant_Analytics/Philosophy_project/Data/DataSEP_Articles_Citations.csv')
df.columns
```

First let me process the body of the articles. I'd love to see how word-document frequcnes look like. Maybe do some cosign-similarity analysis?

```python
t1 = df['body']
```

```python
# Making a term-document matrix

from sklearn.feature_extraction.text import CountVectorizer
token_pattern = r'\b[A-Za-z]+\b'

# I am filtering out any words that show up in the top .05 most frequent percentile
# and the bottom .1 lest frequeny terms 
count_vect = CountVectorizer(stop_words = 'english', token_pattern=token_pattern)

# Make a term document matrix & turn it into a data frame
token_matrix = count_vect.fit_transform(t1)
token_df= pd.DataFrame(token_matrix.toarray(), columns=count_vect.get_feature_names_out())

# Get word final list and table dimentions
word_list = list(token_df.columns)
print(token_df.shape)

# compare filtered data to total
count_vect_loose = CountVectorizer(stop_words = 'english')
ltoken_matrix = count_vect_loose.fit_transform(t1)
ltoken_df= pd.DataFrame(ltoken_matrix.toarray(), columns=count_vect_loose.get_feature_names_out())
print(ltoken_df.shape)

lword_list = list(ltoken_df.columns)
```

```python
def subtract_sets(set1, set2):
  """Subtracts two sets and returns the difference.

  Args:
    set1: The first set.
    set2: The second set.

  Returns:
    The difference of the two sets.
  """

  difference = set()
  for element in set1:
    if element not in set2:
      difference.add(element)
  return difference
```

```python
set1 = set(lword_list)
set2 = set(word_list)

subtract_sets(set1,set2)
```
